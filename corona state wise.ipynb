```
url="https://www.mohfw.gov.in/"
table_list=pd.io.html.read_html(url)
table_list
mf_list=table_list[0]
mf_list.head(4)
```

```
mf_list.columns = ['S.No.','State' , 'Confirmed cases', 	'Cured'	,'Death']
```

```
mf_list.head(3)
```

```
coron19=mf_list.drop(['S.No.'],axis=1)
```

```
coron19.head(31)
```

```
new=coron19.drop(coron19.index[30])
```

```
new.head(30)
```

```
fig = px.bar(new.sort_values('Death', ascending=False),
             x="State", y="Death", color='Death', 
             text='State', orientation='v', title='Death', 
             range_x=[-5,50],
             color_discrete_sequence = [h, c, s, m, e])
             

fig.update_traces(textposition='auto')
fig.show()

```

```
fig = px.bar(new.sort_values('Confirmed cases', ascending=False),
             x="State", y="Confirmed cases", color='Confirmed cases', 
             text='State', orientation='v', title='Confirmed Cases', 
             range_x=[-5,50],
             color_discrete_sequence = [h, c, s, m, e])

fig.update_traces(textposition='inside')

fig.show()
```

```
fig = px.bar(new.sort_values('Cured', ascending=False),
             x="State", y="Cured", color='Cured', 
             text='State', orientation='v', title='Cured', 
             range_x=[-5,50],
             color_discrete_sequence = [h, c, s, m, e])
             

fig.update_traces(textposition='auto')
fig.show()
```
